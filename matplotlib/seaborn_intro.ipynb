```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")  # Options: whitegrid, dark, white, ticks
```

```python
x = np.array([0, 2, 3, 4, 5, 6, 7, 8, 9 ,10, 60])
y = np.sin(x)

sns.lineplot(x=x, y=y)
plt.title('Beautiful Line Plot')
plt.show()
```

```python
x1 = np.linspace(0, 4 * np.pi, 500)
y1 = np.tan(x1)

# plt.plot(x1,y1)
sns.lineplot(x =x1, y = y1)
plt.title('tan graph')

plt.ylim(-10, 10)  # Optional: limit y-axis for better visibility
# plt.grid(True)
# plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 1000) # Creates 1000 points between 0 and 4

y = np.sin(x)

plt.plot(x, y)

plt.xlabel("Angle (radians)")
plt.ylabel("Sine value")
plt.title("Sine Wave Graph")
plt.grid(True)
```

```python
# sns.style.available
sns.axes_style()
```

```python
tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Default (Darkgrid) Style")
plt.show()

# Set the style to 'whitegrid' and plot again
sns.set_style("whitegrid")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Whitegrid Style")
plt.show()

# Set the style to 'ticks' and plot
sns.set_style("darkgrid")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Ticks Style")
plt.show()
```
