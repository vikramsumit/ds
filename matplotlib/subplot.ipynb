```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
y3 = [i ** 3 for i in x]
```

```python
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.title('Double of x')
plt.xlabel('Number')
plt.ylabel('Double of Number')

plt.subplot(1,3,2)
plt.plot(x,y2)
plt.title('Square of x')
plt.xlabel('Number')
plt.ylabel('Square of Number')

plt.subplot(1,3,3)
plt.plot(x,y3)
plt.title('Cube of x')
plt.xlabel('Number')
plt.ylabel('Cube of Number')

plt.tight_layout()
```

```python
# More variations of x
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]

plt.figure(figsize=(8, 6))  # Optional: make it bigger

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('x * 2')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('x squared')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('sqrt(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('10 - x')

plt.tight_layout()
plt.show()
```
