```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris_train.xlsx")
```

```python
# data.head
data.head()
```

```python
# data.tail()
```

```python
x = data.iloc[:, : -1] #rows then column-1
x
```

```python
y = data.iloc[:, -1]
y
```

```python
model = RandomForestClassifier()
```

```python
# x.values
```

```python
# y.values
```

```python
model.fit(x.values,y.values)
# x will give error as it does not contain array
```

```python
predictions = model.predict([[4.9, 2.4, 3.3, 1]])
```
